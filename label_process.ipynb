```python
air_label_dic = pd.DataFrame({
    'air_component':['NO2','PM10','PM2.5','O3'],
    'princple':[
        [0,68,135,201,268,335,401,468,535,601],
        [0,17,34,51,59,67,76,84,92,101],
        [0,12,24,36,42,48,54,59,65,71],
        [0,34,67,101,121,141,161,188,214,241]]
        })
```

```python
air_label_dic
```

```python
def label_maker(comp=[],prcp=[],data=data):
    label_ls=[]
    i=0
    for a in data[comp]:
        lable=10
        for j in range(len(prcp)-1):
            if a >= prcp[j] and a < prcp[j+1]:
                #print(j+1)
                label=j+1
                i=i+1
        label_ls.append(label)    
        #print(a,i,prcp[j],prcp[j+1],label,sep='-----')
    return label_ls
```

```python
data['NO2_label']=label_maker(comp='NO2',prcp=air_label_dic.iloc[0]['princple'])
data['PM10_label']=label_maker(comp='PM10',prcp=air_label_dic.iloc[1]['princple'])
data['PM2.5_label']=label_maker(comp='PM2.5',prcp=air_label_dic.iloc[2]['princple'])
data['O3_label']=label_maker(comp='O3',prcp=air_label_dic.iloc[3]['princple'])
```

```python
data['AQI_IDX']=data[['NO2_label','PM10_label','PM2.5_label','O3_label']].max(axis=1)
```

```python
hour_data=data.groupby(['Year','Month','Date','Hour'],as_index=True).agg(max_AQI=('AQI_IDX','max'),
                                                                         min_AQI=('AQI_IDX','min'),
                                                                         mean_AQI=('AQI_IDX','mean'))
#hour_data.mean_AQI=hour_data.mean_AQI.astype(int)
```
