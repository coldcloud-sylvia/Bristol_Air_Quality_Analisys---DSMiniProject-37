```python
import pandas as pd

traffic = pd.read_csv("fact-traffic-counts.csv", delimiter=';')
```

```python
traffic.head(5)
```

```python
traffic = traffic[['Date Time', 'Hourly Flow', 'Hourly Profile', 'Link', 'geo_point_2d']]
traffic.sort_values('Date Time', ascending = True)
traffic.head(5)
```

```python
# traffic['Hour_sum'] = traffic['Date Time'].apply(lambda x: x.split(':')[2])
# traffic['Hour_0'] = traffic['Hour_sum'].apply(lambda x: x.split('+')[2])
# traffic['Date'] = traffic['Date Time'].apply(lambda x: x.split('+')[0])
traffic['Date'] = traffic['Date Time'].apply(lambda x: x.split('+')[0])
traffic['Year'] = pd.to_datetime(traffic['Date'],format='%Y-%m-%dT%H:%M:%S').dt.year
traffic['Month'] = pd.to_datetime(traffic['Date'],format='%Y-%m-%dT%H:%M:%S').dt.month
traffic['Day'] = pd.to_datetime(traffic['Date'],format='%Y-%m-%dT%H:%M:%S').dt.day
traffic['hour'] = pd.to_datetime(traffic['Date'],format='%Y-%m-%dT%H:%M:%S').dt.hour

traffic['Longitude'] = traffic['geo_point_2d'].apply(lambda x: x.split(',')[0])
traffic['Latitude'] = traffic['geo_point_2d'].apply(lambda x: x.split(',')[1])
traffic = traffic.astype({'Longitude':'float', 'Latitude':'float'})
traffic = traffic[['Year', 'Month', 'Day', 'hour', 'Hourly Flow', 'Hourly Profile', 'Link', 'Longitude', 'Latitude']]
traffic.head(5)
```
