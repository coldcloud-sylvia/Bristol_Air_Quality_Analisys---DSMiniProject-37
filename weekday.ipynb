```python
import pandas as pd

daily_data = pd.read_csv("daily_data.csv")
```

```python
daily_data.head(50)
```

```python
# 循环遍历 Weekday 列中从第 7 行开始的每一行，并将数字 1-7 按顺序循环填写到每一行中
count = 1
for index, row in daily_data.iterrows():
    if index >= 6:  # 第 7 行的索引为 6
        if count > 7:
            count = 1
        daily_data.loc[index, 'Weekday'] = count
        count += 1
```

```python
daily_data.head(5)
```

```python
daily_data['Year'].unique()
```

```python
# 筛选出 Year 列值为 2019 的行，并删除这些行
daily_data = daily_data.drop(daily_data[daily_data['Year'] == 2019].index)
```

```python
daily_data.head(5)
```

```python
result = daily_data.groupby('Weekday').mean()
```

```python
Month = daily_data.groupby('Month').mean()
```

```python
Month.to_csv('Month_avg.csv')
```

```python
df = daily_data
```

```python
# 按照条件筛选数据并求平均值
df1 = df[df['Month'].isin([1, 2, 3])]
avg1 = df1.mean().to_frame().transpose()

df2 = df[df['Month'].isin([4, 5, 6])]
avg2 = df2.mean().to_frame().transpose()

df3 = df[df['Month'].isin([7, 8, 9])]
avg3 = df3.mean().to_frame().transpose()

df4 = df[df['Month'].isin([10, 11, 12])]
avg4 = df4.mean().to_frame().transpose()

# 将每个 DataFrame 转置并连接在一起
result = pd.concat([avg1, avg2, avg3, avg4])
```
